```python
import torch

use_gpu = torch.cuda.is_available()
```

```python
import os
from google.colab import files
import IPython
from TTS.api import TTS

tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=False, gpu=use_gpu)

def clone_voice(tts, text: str, voice: str = None, lan='en'):
  if voice.endswith('.wav'):
    voice = voice.split('.wav')[0]
  if not os.path.exists(f'{voice}.wav'):
    # select only one file
    for i, file_data in enumerate(files.upload().values()):
      with open(f'{voice}.wav', 'wb') as f:
        f.write(file_data)

  tts.tts_to_file(text, speaker_wav=f'{voice}.wav', language=lan, file_path=f'output_{voice}.wav')
  return f'output_{voice}.wav'
```

```python
# Tony Stark - short sentence

text = 'We must destroy all of them and revenge our planet'
voice = 'tony_stark'
voice = clone_voice(tts, text, voice, 'en')
IPython.display.Audio(voice)
```

```python
text = 'Dobbiamo distruggerli tutti e rivendicare il nostro pianeta'  # Italian
voice = 'tony_stark'
voice = clone_voice(tts, text, voice, 'en')
IPython.display.Audio(voice)
```

```python
text = 'Nous devons tous les détruire et venger notre planète'  # Italian
voice = 'tony_stark'
voice = clone_voice(tts, text, voice, 'fr-fr')
IPython.display.Audio(voice)
```
