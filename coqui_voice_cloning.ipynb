```python
from TTS.api import TTS
import IPython

tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=False, gpu=True)
text = 'One of the key advantages of AI is its ability to process and analyze vast amounts of data, which can help us gain insights, identify patterns, and make predictions. This has applications across many industries, including healthcare, finance, manufacturing, and transportation.'
tts.tts_to_file(text, speaker_wav="weapon.wav", language='en', file_path="output_tony.wav")
IPython.display.Audio('output_tony.wav')
```
